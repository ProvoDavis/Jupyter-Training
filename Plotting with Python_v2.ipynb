### Useful data types

```python
#Store stuff in lists!

#these are arrays that can contain any datatype. Very powerful for creating collections of things to iterate over in a specific order

l = [] #empty list

#examples - print them out

l1 = [0,1,2,3,4,5]

l2 = ['A','B','C','D','E','F','G','H']

l3 = [[1,2,3],[4,5,6],[7,8,9]]

l4 = ['A',1,'B',2,'C',3, [1,2,3]]


```

```python
#Store objects in sets!

#sets are collections of unique items. They are useful to do operations like intersections


s = ()

#or similar

s = set()

s1 = set([0,1,2,3,4,5,6,'A','B'])
s2 = set(l2)

 
print(s1.intersection(s2)) #What are you seeing here?
```

```python
#Store key, value pairs in dictionaries!

#Super useful data structure! This allows you to find variables by identifiers / keys

d = {} 

#or similar

d = dict()
```

```python
#dictionaries are useful, and can be iterated over


# Dictionaries can be initialized by specifying key-value pairs in the form: {key:value, ...} 

d = {'a':1,'b':2,'c':3,'d':4,'e':5,'f':5}

print("The dictionary", d)
print('has key and value pairs:')

for key in d:
    print(key,d[key])

```

```python
#another nice trick are enumerations

#enumerate returns both indexes and data values**

d = {'a':1,'b':2,'c':3,'d':4,'e':5,'f':6}

print("The dictionary", d)
print('has index, key, and value pairs:')

for i,key in enumerate(d):
    print(i,key,d[key])
```
